```python
query = {
    "query": {
        "bool": {
            "must": [
                {
                    "term": {"Region.keyword": "Midwest"}
                },
                {
                    "bool": {
                        "should": [
                            {"term": {"Short Month.keyword": "Jan"}}
                        ]
                    }
                }
            ]
        }
    }
}
```

```python
#Dispalying the employees who belong to midwest and joined in jan month
res = es.search(
  index = 'metafile',
  size = 100,
  body = query)
print(res)
```

## Summarise the number of emp joined and hickes granted based on month

```python
#Dispalying employees count based on month
try:
    print(es.search(index="metafile", body={"size":0,"aggs":{"by_month":{"terms":{"field":"Short Month.keyword"}}}}))
except:
    print("records not found")
```

```python
#Dispalying employees count based on month
es.search(index="metafile", body={"size":0,"aggs":{"by_month":{"terms":{"field":"Hike.keyword"}}}})
```

```python
#Extarctinfg the deatils based on months in ascending order
es.search(index="metafile", body={
                              "aggs": {
                                "genres": {
                                  "terms": {
                                    "field": "Short Month.keyword",
                                    "order": { "_count": "asc" }
                                  }
                                }
                              }
                            })
```

```python
es.search(index="metafile", body={"size":0,"aggs":{"distinct_months":{"terms":{"field":"Short Month.keyword"}}},
                                 "aggs":{
                                     "by_hike":{
                                         "terms":{
                                             "field":"Hike.keyword"
                                         }
                                     }
                                 }
                            })
```
